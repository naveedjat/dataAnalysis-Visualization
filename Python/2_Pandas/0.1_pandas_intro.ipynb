# Introduction to Pandas

## Series - 1D Labeled Array

```python
import pandas as pd

s = pd.Series([10, 20, 30, 40])
print(s)
```

```python
s = pd.Series([10, 20, 30], index=["a", "b", "c"])
print(s)
```

```python
data = {
    "name": ["Naveed", "Ziah", "Ashwariya"],
    "age": [19,19,19],
    "city": ["Samaro", "Hyderabad", "Lahore"]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.index = ["a", "b", "c"]
df.columns = ["Name", "Age", "City"]
print(df)
```
