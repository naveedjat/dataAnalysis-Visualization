# Matplotlib 


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot') # it changes the plots style like ggplot
```

```python
#ploting an simple function - 2D line plot
price=[40000,41000,47000,59000,48000,49000,40000]
years=[2019,2020,2021,2022,2023,2024,2025]
plt.plot(years,price)
```

```python
batsman = pd.read_csv('sharma-kohli.csv')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# labels title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title("Rohit Sharma Vs Virat Kohli Career Comparison")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
# color hex and line(style and width) and marker(size)
plt.figure(facecolor='orange')  # Changes entire figure background

plt.plot(batsman['index'],batsman['V Kohli'] , color = 'green',linestyle='dashdot',linewidth=2)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashed',linewidth=3)
plt.title("Rohit Sharma Vs Virat Kohli Career Comparison",color="Purple")
plt.xlabel("Season",color="blue")
plt.ylabel("Runs Scored",color="brown")
plt.savefig('plot.webp')  # it saves the plot in any format like jpg , png,webp,and jpeg
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],marker='s',markersize=7)
plt.plot(batsman['index'],batsman['RG Sharma'],marker='o',markersize=9)
plt.title("Rohit Sharma Vs Virat Kohli Career Comparison")
plt.xlabel("Season")
plt.ylabel("Runs Scored")


#Markersize increase markers size according to teh requirements
# markersize = n        



#Markers in matplotlib
# 'o'	Circle	
# '*'	Star	
# '.'	Point	
# ','	Pixel	
# 'x'	X	
# 'X'	X (filled)	
# '+'	Plus	
# 'P'	Plus (filled)	
# 's'	Square	
# 'D'	Diamond	
# 'd'	Diamond (thin)	
# 'p'	Pentagon	
# 'H'	Hexagon	
# 'h'	Hexagon	
# 'v'	Triangle Down	
# '^'	Triangle Up	
# '<'	Triangle Left	
# '>'	Triangle Right	
# '1'	Tri Down	
# '2'	Tri Up	
# '3'	Tri Left	
# '4'	Tri Right	
# '|'	Vline	
# '_'	Hline
```

```python
# legend in matplotlib - for labeling the 

plt.plot(batsman['index'],batsman['V Kohli'],marker='s',markersize=7,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],marker='o',markersize=9,label='Rohit')
plt.title("Rohit Sharma Vs Virat Kohli Career Comparison")
plt.xlabel("Season")
plt.ylabel("Runs Scored")


plt.legend(fontsize='7') #it is used to change the fontsize in the matplotlib
# plt.legend(loc='best') 
plt.legend(loc='lower right') # it is used to locate the legend in the plot example: upper left , upper right, lower right , lower left ,
```

```python
price=[10000,21000,7000,29000,30000,42000,700000]
years=[2019,2020,2021,2022,2023,2024,2025]
plt.plot(years,price)

#it trims the values on x and y axis called limiting axis
plt.ylim(0,75000) #it limits the value on y axis so that outlier might not affect it
plt.xlim(2019,2024) #it limits the value on x axis so that outlier might not affect it
```

```python
#grid 
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.grid()



plt.minorticks_on() #it creates the minor ticks on the Numerical data



plt.grid(alpha=0.4) # it is used to change the transparency of the grid
plt.grid(color='gray', linestyle='--', linewidth=0.5) #grid style


plt.show()  # it is used to show the plot in the every codeeditor like , vim, vscode , pycharm where it is compulsory to use it .
```
